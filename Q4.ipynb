```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, unix_timestamp, max as spark_max
```

```python
spark = SparkSession.builder \
    .master("local[*]") \
    .appName('homework') \
    .getOrCreate()
```

```python
df = spark.read.parquet("yellow_tripdata_2025-11.parquet")
```

```python
df.show()
```

```python
df = df.withColumn("trip_duration_hours", (unix_timestamp(col("tpep_dropoff_datetime")) -
     unix_timestamp(col("tpep_pickup_datetime"))) / 3600
)
```

```python
longest_trip = df.select(
    spark_max("trip_duration_hours")
).collect()[0][0]
```

```python
print("Longest trip duration:", longest_trip)
```
